```python
final = pd.read_csv(r"C:\Users\leila\Downloads\pfaproject\house-price-prediction\data\cleaned data\finaltrain.csv")
```

```python
# pc1, pc2, one hot encoded
desired_columns = [
    'OverallQual', 'GarageCars', 'GarageArea', 'ExterQual', 'GarageFinish',
    'KitchenQual', 'BsmtQual', 'YearBuilt', 'TotalBsmtSF', 'GrLivArea',
    'TotRmsAbvGrd', '2ndFlrSF', 'BedroomAbvGr', 'BsmtFinType1',
    'BsmtFullBath', 'BsmtFinSF1', 'BsmtUnfSF', 'FullBath', 'KitchenAbvGr', 'SalePrice'
]
one_hot_cols = [
    col for col in final.columns
    if final[col].dropna().isin([True, False]).all()
]
```

```python
# Combine desired columns with one-hot encoded columns
columns_to_keep = desired_columns + one_hot_cols

# Ensure 'Id' is not in the list
columns_to_keep = [col for col in columns_to_keep if col != 'Id']

# Filter the DataFrame
final_filtered = final[columns_to_keep]
```

```python
setup(data=final_filtered, target='SalePrice', session_id=123)
```
