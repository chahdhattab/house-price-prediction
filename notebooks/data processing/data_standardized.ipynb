# TRAIN STANDARDISATION

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv(r"C:\Users\leila\Downloads\pfaproject\house-price-prediction\notebooks\data processing\cleantrain.csv")
```

```python
df.head()
```

```python
# Sélection juste des colonnes numériques
#numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
numeric_cols = [
    col for col in df.select_dtypes(include=['int64', 'float64']).columns 
    if col not in ['Id', 'SalePrice']
]
```

```python
scaler = StandardScaler()
```

```python
# Appliquer la standardisation
df_standardized = df.copy()
df_standardized[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```

```python
#Sauvegarder le résultat
df_standardized.to_csv("cleantrain_standardized.csv", index=False)
```

```python
finaltrain_df = pd.read_csv("cleantrain_standardized.csv", na_filter = False)
```

```python
finaltrain_df.to_csv("finaltrain.csv", index = False)
```
